```python
def get_relevant_documents(prompt, kb_embeddings, top_k=3):
    
    result = {}
    prompt_embedding = model.encode([prompt])
    
    cosine = cosine_similarity(prompt_embedding, kb_embeddings)[0]
    print("Distâncias por cosseno:", cosine)
    top_k_indices = np.argsort(-cosine)
    print("Índices dos documentos (cosseno:)", top_k_indices)
    top_k_indices = top_k_indices[-top_k:]

    result["cosseno"] = {
        "distances": cosine,
        "indices": top_k_indices,
        "documents": [knowledge_base[i] for i in top_k_indices]
    }

    euclidean = euclidean_distances(prompt_embedding, kb_embeddings)
    print("Distâncias euclidianas:", euclidean)
    euclidean = euclidean[0]
    top_k_indices = np.argsort(euclidean)
    print("Índices dos documentos (euclidiana):", top_k_indices)
    top_k_indices = top_k_indices[:top_k]

    manhattan = manhattan_distances(prompt_embedding, kb_embeddings)
    print("Distâncias de Manhattan:", manhattan)
    manhattan = manhattan[0]
    top_k_indices = np.argsort(manhattan)
    print("Índices dos documentos (Manhattan):", top_k_indices)
    top_k_indices = top_k_indices[:top_k]

    pairwise = pairwise_distances(prompt_embedding, kb_embeddings, metric='cosine')
    print("Distâncias par-a-par (pairwise):", pairwise)
    pairwise = pairwise[0]
    top_k_indices = np.argsort(pairwise)
    print("Índices dos documentos (pairwise):", top_k_indices)
    top_k_indices = top_k_indices[:top_k]
    
    return [knowledge_base[i] for i in top_k_indices]

def print_relevant_documents(prompt, kb_embeddings, top_k=3):
    relevant_docs = get_relevant_documents(prompt, kb_embeddings, top_k)
    for i, doc in enumerate(relevant_docs, 1):
        print(f"Documento relevante {i}: {doc['title']}")


prompts = [
    {"correct": 0, "prompt": "Conte-me a história de uma menina que visita sua avó na floresta e encontra um lobo."},
    {"correct": 1, "prompt": "Quais são os desafios enfrentados por três irmãos que constroem suas próprias casas?"},
    {"correct": 2, "prompt": "Fale sobre uma princesa que é protegida por sete anões."},
    {"correct": 3, "prompt": "Qual é a história de uma jovem que é maltratada por sua madrasta e irmãs, mas acaba encontrando o amor verdadeiro?"}
]

for model_info in models:
    model = model_info["model"]
    kb_embeddings = model_info["embedding"]
    print(f"\n=== Resultados para o modelo: {model_info['name']} ===\n")
    for prompt in prompts:
        print(f"\nPrompt: {prompt['prompt']}\n")
        print_relevant_documents(prompt, kb_embeddings, top_k=2)

```
